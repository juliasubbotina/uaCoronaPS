```python
import os, sys
import MDAnalysis as mda
import nglview as nv
```

```python
#Dictionaries
labelDic={  "1hzh":"IGH",
            "1ao6":"HSA",
            "3k2s":"APO-AI",
            "3ghg":"FG",
            "2a73":"C3"}

beadDicRGB={ "PS": [0.86, 0.86, 0.86],
            "NH3": [0.0, 0.5, 1.0],
            "NH2": [0.0, 0.5, 1.0],
           "COOH": [1.0, 0.13, 0.32],
            "COO": [1.0, 0.13, 0.32]}

beadDicHex={ "PS": "#DCDCDC",
            "NH3": "#007FFF",
            "NH2": "#007FFF",
           "COOH": "#FF2052",
            "COO": "#FF2052"}
```

```python
wd=os. getcwd() 
pdb_folder = wd+"/pdbs/"
```

```python
pdbs=[]
for pr in os.listdir(pdb_folder):
    if pr.endswith(".pdb"):
            pdbs.append(pr)
pdbs
```

```python
i=2
```

```python
system=pdbs[i].rsplit(".")[0]
system
```

```python
radii=0.38 # shell beads
radii_core=float(system.rsplit("_")[1].replace("nm",""))-radii
core="PS"
```

```python
protein=system.rsplit("_")[2]
protein
```

```python
u=mda.Universe('{}{}.pdb'.format(pdb_folder,system))
```

# Visualize NP-protein complex

```python
view=nv.show_mdanalysis(u)
view._remote_call('setSize', target='Widget', args=['600px','900px'])
view.clear_representations()
view.add_representation(repr_type='cartoon', selection='protein')
for bds in list(beadDicHex.keys()):
    view.add_representation(repr_type='spacefill', selection=bds,  opacity=0.6, color=beadDicHex[bds], radius=(radii*1.0)*10)
central = view.shape
central.add_sphere([ 0.0, 0.0, 0.0 ], beadDicRGB[core], radii_core*10)
view
```
