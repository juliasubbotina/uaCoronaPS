```python
import os, sys
import MDAnalysis as mda
import nglview as nv
```

```python
#Dictionaries
labelDic={  "1hzh":"IGH",
            "1ao6":"HSA",
            "3k2s":"APO-AI",
            "3ghg":"FG",
            "2a73":"C3"}

beadDicRGB={ "PS": [0.86, 0.86, 0.86],
            "NH3": [0.0, 0.5, 1.0],
            "NH2": [0.0, 0.5, 1.0],
           "COOH": [1.0, 0.13, 0.32],
            "COO": [1.0, 0.13, 0.32]}

beadDicHex={ "PS": "#DCDCDC",
            "NH3": "#007FFF",
            "NH2": "#007FFF",
           "COOH": "#FF2052",
            "COO": "#FF2052"}
```

```python
wd=os. getcwd() 
pdb_folder = wd+"/pdbs/"
```

```python
pdbs=[]
for pr in os.listdir(pdb_folder):
    if pr.endswith(".pdb"):
            pdbs.append(pr)
```

```python
def find_keys_by_value(d, value):
    return [key for key, val in d.items() if val == value][0]
```

```python
i_p=0 # change from 0 to 4 to go through all proteins
identifier=list(labelDic.values())[i_p]
material=["psps","pscooh","psnh2"][0] #change from 0 to 2 to go through all NPs
ps=material[2:]
R=5.0
core="PS"
pdbId=find_keys_by_value(labelDic, identifier)
pdbFile ="{}_{}nm_{}_CG.pdb".format(material,int(R),pdbId)
radii=0.38 # shell beads
radii_core=R-radii
combo=mda.Universe('{}{}'.format(pdb_folder,pdbFile))
```

```python
view=nv.show_mdanalysis(combo)
view._remote_call('setSize', target='Widget', args=['600px','900px'])
view.clear_representations()
view.add_representation(repr_type='cartoon', selection='protein')



for bds in list(beadDicHex.keys()):
    view.add_representation(repr_type='spacefill', selection=bds,  opacity=0.6, color=beadDicHex[bds], radius=(radii*1.0)*10)

central = view.shape
central.add_sphere([ 0.0, 0.0, 0.0 ], beadDicRGB[core], radii_core*10)




view
```
