```python
#Convert to one letter code
def convert(seq):
    d = {'CYS': 'C', 'ASP': 'D', 'SER': 'S', 'GLN': 'Q', 'LYS': 'K',
     'ILE': 'I', 'PRO': 'P', 'THR': 'T', 'PHE': 'F', 'ASN': 'N', 
     'GLY': 'G', 'HIS': 'HP', 'LEU': 'L', 'ARG': 'R', 'TRP': 'W', 
     'ALA': 'A', 'VAL':'V', 'GLU': 'E', 'TYR': 'Y', 'MET': 'M', 
    'ASPP': 'DP', 'HIS': 'HP', 'HID': 'HD', 'HIE': 'HE', 'GLUP': 'EP' , 'CYM': 'CM'}
    
    single_seq=''
    single_seq += d[seq[:]]
    return single_seq
```

```python
dic_js_to_UA =        {"GLY":"GLY","ALA":"ALA","ARG":"ARG","ASN":"ASN",
                      "ASP":"ASP","CYS":"CYS","GLN":"GLN","GLU":"GLU",
                      "ILE" :"ILE","LEU":"LEU","LYS":"LYS",
                      "MET":"MET","PHE" :"PHE","PRO":"PRO","SER":"SER", 
                      "THR":"THR","TRP" :"TRP","TYR":"TYR","VAL":"VAL",
                      "HSD":"HID","HSE" :"HIE","HSP":"HIP", "HIS":"HIS",
                      "CYM":"CYM","ASPP":"AAN","GLUP":"GAN",
                      "BGLC":"DGL","AMAN":"MAN","AFUC":"FUC", #carbs
                      "BGLCNA":"NGL","BGALNA":"NGA", #N-glycolization
                      "NC4":"CHL", "MAMM":"ETA",      #lipids heads
                      "MAS":"EST","DMEP":"PHO", #lipids phosphotidyl
                      #"BTE2":"BTE", #extra  buten,
                      # "CHL1":"CHO" #extra   cholesterol
                      # "CHOL":"CHL" #extra   full choline, not CG 
                      }
```

```python
def find_key_by_value(d, value):
    for k, v in d.items():
        if v == value:
            return k
    return v
```

```python
find_key_by_value(dic_js_to_UA,"AAN")
```

```python
def plot_pmf(location, ax=None, i=97,  sample=mol_print): 
    ax = ax or plt.gca()
    ncolor=len(sample)
    color = iter(cm.turbo(np.linspace(0, 1, ncolor)))
    
    for n, acid in enumerate(sample):
        #charge=dic_sca[acid]['charge']
#         if acid == 'HID' :
#             acid = 'HSD'
#         if acid == 'HIE' :
#             acid = 'HSE'
#         if acid == 'HIP' :
#             acid = 'HSP'            
#         if acid == 'HIS' :
#             acid = 'HIP'  
        if acid == 'HSD' :
            acid = 'HID'
c = next(color)
        
        harmonised_acid=find_key_by_value(dic_js_to_UA, acid)
        print(harmonised_acid)
        pmf=location+harmonised_acid + '.dat'
        data = genfromtxt(pmf, delimiter=',')
        offset = 0
        data[:,1] -= offset
        minE=np.min(data[:,1])
#         if acid == 'HIS':
#             aa_label = 'HIP'

#         else:
        aa_label = acid
        #labelAA=aa_label+(" ({}, {:.2f})").format(charge, minE)
        labelAA=aa_label
        ax.plot( data[:,0], data[:,1], label = labelAA, c=c,  linewidth=1.5,  rasterized=True) # original data from g(r) in kbT already
        ax.plot([1.2,0],[0,0], 'r--', lw=1.5,  rasterized=True)
        ax.set(xlim=(0.0, 1.5), ylim=(-2.0, 4.0), xlabel="SSD (nm)", ylabel=r"$U_{s}$ (k$_{B}$T)",  rasterized=True)

```

```python
location=os.getcwd()+"/"
decor=location.split("/")[3].split("_")[0]
my_dpi=300
w_pxl=1200
h_pxl=1000
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(w_pxl/my_dpi, h_pxl/my_dpi), dpi=my_dpi )
fig.subplots_adjust(left=0.1, right=0.9, wspace=0.3, top=0.9, bottom=0.1)
fig.tight_layout(pad=0.25)
ax1.imshow(img,  aspect='auto', extent=[1.0, 1.5, -2.5, 0.0])

plot_pmf(location, ax1, i=57, sample=mol_print)
labels_handles = {label: handle for ax in fig.axes for handle, label in zip(*ax.get_legend_handles_labels())}

fig.legend(
  labels_handles.values(),
  labels_handles.keys(),
  loc="upper center", ncol=3 ,  fontsize=8, frameon=False, #edgecolor='black', title=r"$\bf{SCAs}$"+'\n'
  bbox_to_anchor=(0.70, 0.95),
  bbox_transform=plt.gcf().transFigure, prop={'size': 6}
)


figExt=["eps", "png", "pdf"]
ids=-1
fig.savefig(decor+'_pmfs.'+figExt[ids], format=figExt[ids], dpi=my_dpi, bbox_inches='tight')
```
